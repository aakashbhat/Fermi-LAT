```python
import numpy as np
import pandas as pd
from astropy.io import fits
```

```python
folder = '../ML_3FGL_paper/data/python files/catas'
folder_out = 'data'
cat = '4FGL'

cat_fn = '%s/%s_unassoc_catalog_unweighted_col_names_PC.csv' % (folder, cat)
unassoc_cat = pd.read_csv(cat_fn)
cat_fn = '%s/%s_assoc_catalog_unweighted_col_names_PC.csv' % (folder, cat)
assoc_cat = pd.read_csv(cat_fn)
out_fn = '%s/%s_prob_catalog.csv' % (folder_out, cat)

if cat == '3FGL':
    fn = '../PS_comparison/gll_psc_v16.fit' # the last release of 3FGL
else:
    fn = '../PS_comparison/gll_psc_v22.fit' # 4FGL release from 27 May 2020
hdu = fits.open(fn)
index = np.array(hdu[1].data.field('Source_Name'), dtype=str)
key = 'GLON'
data = {key:np.array(hdu[1].data.field(key), dtype=float)}
df_fgl = pd.DataFrame(data=data, index=index)
df_fgl.index = [st.strip() for st in df_fgl.index]
```

```python
merged_cat = assoc_cat.merge(unassoc_cat, how='outer').set_index('Source_Name_%s' % cat)
for name in ['BDT_P','NN_P', 'LR_P', 'RF_P', 'GLON']:
    merged_cat.pop(name)
```

```python
df_fgl = df_fgl.loc[merged_cat.index]
```

```python
merged_cat = merged_cat.join(df_fgl)
```

```python
print('save to file:')
print(out_fn)
merged_cat.to_csv(out_fn, index=True, float_format='%.9g')
```
